# Importing The Libraries

```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
```

# Read the Image

```python
img = cv2.imread("dog.jpg")
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
display(Image.fromarray(img))
```

# Converting the Original Image into Grayscale

```python
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grayImg))
```

# Inverting the Grayscale Image

```python
inv_grayImg = 255 - grayImg
```

```python
inv_grayImg
```

```python
display(Image.fromarray(inv_grayImg))
```

# Blur the Image

```python
blur_img = cv2.GaussianBlur(inv_grayImg, (35,35), 0)
```

```python
display(Image.fromarray(blur_img))
```

# Mixing the grayscale image and inverted blur image

```python
inv_blurImg = 255 - blur_img
```

```python
inv_blurImg
```

```python
display(Image.fromarray(inv_blurImg))
```

```python
def mix(grayImg, blur_img):
    return cv2.divide(grayImg, inv_blurImg, scale = 210.0)
```

```python
pencil_sketch = mix(grayImg, inv_blurImg)
display(Image.fromarray(pencil_sketch))
```
